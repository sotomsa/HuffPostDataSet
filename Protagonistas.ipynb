```python
import pandas as pd
import numpy as np
import requests
from pprint import pprint
import os
import json
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df_org = pd.read_json('Data/News_Category_Dataset_v2.json',lines=True)
df_org['id'] = range(len(df_org)) 
df_org
```

```python
sample = df_org[['id','short_description','category']].sample(100)
sample =  sample.rename(columns={'short_description':'text'})
sample.to_pickle('Data/sample.pkl')
```

```python
cognitive_service_key = 'https://prueba-servicios-cognitivos.cognitiveservices.azure.com/'
entities_url = endpoint + "/text/analytics/v3.0/entities/recognition/general"
headers = {"Ocp-Apim-Subscription-Key": subscription_key}
```

```python
import math
ndocs = 5
nrows = len(sample)
nrequest = math.ceil(nrows/5)
jsons = []
for i in range(nrequest):
    start = i*ndocs
    end = start + ndocs
    sample_json = {'documents': json.loads(sample.iloc[start:end].to_json(orient='records'))}
    response = requests.post(entities_url, headers=headers, json=sample_json)
    entities = response.json()
    jsons.extend(entities['documents'])

entitiesAzure = pd.DataFrame(jsons)
entitiesAzure
```

```python
entitiesAzure['id'] = pd.to_numeric(entitiesAzure['id'])
data = sample.merge(entitiesAzure, on='id')
data
```

```python
data_new = data.explode('entities').dropna(axis=0,how='any')
data_new = data_new.rename(columns={'text':'text_long',
                                    'category':'category_news'})
p1 = data_new.reset_index(drop=True)
p2 = pd.json_normalize(data_new['entities']).reset_index(drop=True)
final = pd.concat([p1,p2], axis = 1)
entitiesAzure.to_pickle('Data/enitites.pkl')
final
```
